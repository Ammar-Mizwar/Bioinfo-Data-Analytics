```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#a 2-by-2 array containing the numbers 0–3
D = np.arange(4).reshape((2,2))
print("2-by-2 array containing the numbers 0–3: \n\n", D, "\n")

#Cube every element in the array
E = D**3
print("Cubed every elements: \n\n", E, "\n")

#Add 7 to every element in the array
F = D+7
print("Added 7 to every elements: \n\n", F, "\n")

#Multiply every element of the array by 2
G = D*2
print("Multiplied every elements by 2: \n\n", G, "\n")
```

```python
# 10-by-10 array containing the number 4 in each position
H = np.full((10,10), 4)
print("10-by-10 array containing the number 4 in each position: \n\n", H, "\n")

#10-by-10 array containing the numbers 1 to 100 in ascending order
I = np.arange(1,101).reshape((10,10))
print("10-by-10 array containing the numbers 1 to 100 in ascending order: \n\n", I, "\n")

#H multiplied by I
J = H*I
print("Multiplied both together: \n\n", J, "\n")
```

```python
# List of lists
K = np.arange(10, -1, -2)  # Down from 10 to 0 (inclusive)
L = np.arange(0, 11, 2)    # Up from 0 to 10 (inclusive)
# Combine into a list of lists and convert to array
M = [K.tolist(), L.tolist()]
print("List of lists: \n\n", M, "\n")

# Turn K and L to array
N = np.array(M)
print("Array from lists above: \n\n", N, "\n")
```

```python
# Change to 1D using flatten
O = N.flatten()
O[1] = 10  # <- Change second element to 10
print("\nModified flattened COPY:\n\n", O, "\n")
print("Original N after .flatten():\n\n", N, "\n")

# Change to 1D using ravel
P = N.ravel()
P[1] = 10  # <- Change second element to 10
print("\nModified flattened COPY:\n\n", P, "\n")
print("Original N after .ravel():\n\n", N, "\n")

# Explanation
print("\nExplanation:\n")
print("The flatten() method returns a new, independent 1D copy of the original array.\n")
print("Any modifications to the flattened copy do not affect the original array.\n")
print("On the other hand, ravel() returns a flattened view (not a copy).\n")
print("Meaning any changes made to the raveled version will directly reflect in the original array.\n")
```

```python
def show(title, arr):
        print(f"{title}:\n\n{arr}\n")

# 3-by-5 array of numbers 1 to 15
U = np.arange(1, 16).reshape((3, 5))

show("Original 3x5 array U", U)  # Print full array
show("Row 2 (Positive Indexing)", U[1])                            # Select row 2 (index 1)
show("Row 2 (Negative Indexing)", U[-2])                           # Select row 2 using negative index (-2)
show("Column 4 (Positive Indexing)", U[:, 3])                      # Select column 4 (index 3)
show("Column 4 (Negative Indexing)", U[:, -2])                     # Select column 4 using negative index (-2)
show("Columns 1–3 (Positive Indexing)", U[:, 0:3])                 # Select columns 1–3 (index 0 to 2)
show("Columns 1–3 (Negative Indexing)", U[:, -5:-2])               # Same columns using negative indices
show("Element at row 1, column 4 (Positive Indexing)", U[0, 3])    # Element at row 1, col 4 (index [0,3])
show("Element at row 1, column 4 (Negative Indexing)", U[-3, -2])  # Same element using negative indices
```

```python
array1 = np.array([[0, 2], [4, 6]])
array2 = np.array([[1, 3], [5, 7]])

array3 = np.concatenate((array1, array2), axis=1)           #For axis parameter, 0 = vertical stacking, 1 = horizontal stacking
show("Horizontal stacking of array1 and array2", array3)    #Using the defined 'show' function

array4, array5 = np.vsplit(array3, 2)                       #2 means split into 2 equal parts (must have even rows)
show("array4 (Top Row)", array4)
show("array5 (Bottom Row)", array5)

array6 = np.concatenate((np.sort(array4), 
                         np.sort(array5)), axis=1)          #Sort both array4 and array5 and use horizontal stacking
show("Sorted and stacked (horizontally) array6", array6)  

array7 = np.concatenate((array6, array6 * 10), axis=0)      #Vertically stacked aray6 and product of array6 with 10
show("Array 7", array7)  
```

```python
import copy

dictionary = {'Sophia': [97, 88]}     # Original dictionary

#-----SHALLOW COPY-----
shallow1 = dict(dictionary)           # Shallow Copy using dict() constructor
shallow2 = dictionary.copy()          # Shallow Copy using copy() method of dict
dictionary['Sophia'][0] = 100         # Modify the inner list of the original

# Display results after shallow copy change
print("Original dictionary:\n", dictionary)
print("Shallow copy 1 (dict()):\n", shallow1)
print("Shallow copy 2 (.copy()):\n", shallow2)

#-----DEEP COPY-----
deep_copy = copy.deepcopy(dictionary) # Using deepcopy from the copy module
dictionary['Sophia'][1] = 50          # Modify original again

# Display after deep copy to prove separation
print("\nAfter deepcopy + second modification:")
print("Original dictionary:\n", dictionary)
print("Deep copy:\n", deep_copy)

#Showing each memory address
print("\nMemory addresses of inner lists:")
print("Original:", id(dictionary['Sophia']))
print("Shallow1:", id(shallow1['Sophia']))
print("Shallow2:", id(shallow2['Sophia']))
print("DeepCopy:", id(deep_copy['Sophia']))
```

```python
series_1 = pd.Series([7, 11, 13, 17])                       #Creating a series
series_2 = pd.Series([100] * 5)                             #Creating a series with repeating 100s
series_3 = pd.Series(np.random.randint(0, 100, size = 20))  #Creating a series with repeating 100s
series_4 = pd.Series([98.6, 98.9, 100.2, 97.9],             #Creating a series with index
                     index=['Julie', 'Charlie', 'Sam', 'Andrea'])
dict_s4 = series_4.to_dict()                                #Make series_4 into dictionary
series_5 =pd.Series(dict_s4)                                #Innitialised from dict_s4

def show_pd(title, panda):                                  #Creating 'show_pd" function to ease print() 
        print(f"{title}:\n\n{panda}\n")

show_pd("Series 1: Prime Numbers", series_1)
show_pd("Series 2: Full of 100s", series_2)
show_pd("Series 3: Randoms", series_3)
print("Series 3 Descriptive Statistics:\n\n",
      series_3.describe(), "\n")                            #Print descriptive statistics of the series
show_pd("Series 4: Temperature", series_4)
print("Dictionary from series 4:\n\n", dict_s4,"\n")
show_pd("Series 5: Innitialise from dictionary", series_5)
```

```python
temp_dict = {                                   #Created dictionary of 3 persons with 3 temperatures
    "Maxines": [37, 36, 36.5],
    "James": [35.6, 37, 35],
    "Amanda": [41, 42, 43]
}

df_temp = pd.DataFrame.from_dict(temp_dict,     #Convert temp_dict to dataframe with indices
                                 orient='index', 
                                 columns=["Morning", 
                                          "Afternoon", 
                                          "Evening"])
show_pd("Dataframe from dictionary:", df_temp)
show("Maxines' temperature:", 
     df_temp.loc[["Maxines"]])                  #Indexing only Maxines' data
show("Morning temperature:", 
     df_temp.T.loc[["Morning"]])                #Indexing only Morning data
show("Morning and Evening temperature:", 
     df_temp.T.loc[["Morning", "Evening"]])     #Indexing only Morning and Evening data
show("Maxines and Amanda's temperature:", 
     df_temp.loc[["Maxines", "Amanda"]])        #Indexing only Maxines and Amanda's data
show("Maxines and Amanda's temperature:", 
     df_temp.loc[["Maxines", "Amanda"], ["Morning", "Afternoon"]])    #Indexing only Maxines and Amanda's data in Morning and Afternoon
show("Descriptive Statistics: ",
     df_temp[["Morning", "Afternoon", "Evening"]].describe())         #Generate the desc. stats
show("Transpose:", df_temp.T)                                         #Transpose
show("Temperature DataFrame (Columns Alphabetically Sorted)",         #Sort the column
    df_temp[sorted(df_temp.columns)])
```

```python
#-----LINE PLOT-----
ax = df_temp.T.plot(
    marker = 'x', 
    title = "Temperature Trends", 
    fontsize = 12    #Font size for ticks
)
ax.set_title("Temperature Trends", 
             fontsize = 12, 
             fontweight = 'bold',    #Bold title
             fontname = 'Times New Roman'
            )
ax.set_ylabel("Temperature (°C)",    #label for y-axis
              fontsize = 12, 
              fontname = 'Times New Roman'
             )
ax.set_xlabel("Time of Day",         #label for x-axis
              fontsize = 12, 
              fontname = 'Times New Roman'
             )
ax.legend(title = "Patients",        #label for legends
          prop = {'family': 'Times New Roman', 
                'size': 12}
         )
ax.grid(True)

for label in (ax.get_xticklabels() + ax.get_yticklabels()): 
    #ax.get_xticklabels() = all the labels on the x-axis
    #ax.get_yticklabels() = all the labels on the y-axis
    label.set_fontname('Times New Roman')
    
plt.tight_layout()
plt.show()


#-----BOX PLOT-----
ax = df_temp.T.plot(
    kind = 'box', 
    color = 'black',
    title = "Temperature Trends", 
    fontsize = 12    #Font size for ticks
)
ax.set_title("Temperature Trends", 
             fontsize = 12, 
             fontweight = 'bold',    #Bold title
             fontname = 'Times New Roman'
            )
ax.set_ylabel("Temperature (°C)",    #label for y-axis
              fontsize = 12, 
              fontname = 'Times New Roman'
             )
ax.set_xlabel("Time of Day",         #label for x-axis
              fontsize = 12, 
              fontname = 'Times New Roman'
             )
ax.grid(True)

for label in (ax.get_xticklabels() + ax.get_yticklabels()): 
    #ax.get_xticklabels() = all the labels on the x-axis
    #ax.get_yticklabels() = all the labels on the y-axis
    label.set_fontname('Times New Roman')
    
plt.tight_layout()
plt.show()

#-----BAR-----
ax = df_temp["Afternoon"].plot(
    kind = 'bar',
    color = 'gray',
    title = "Temperature Trends", 
    fontsize = 12    #Font size for ticks
)
ax.set_title("Temperature Trends", 
             fontsize = 12, 
             fontweight = 'bold',    #Bold title
             fontname = 'Times New Roman'
            )
ax.set_ylabel("Temperature (°C)",    #label for y-axis
              fontsize = 12, 
              fontname = 'Times New Roman'
             )
ax.set_xlabel("Time of Day",         #label for x-axis
              fontsize = 12, 
              fontname = 'Times New Roman'
             )
ax.grid(True)

for label in (ax.get_xticklabels() + ax.get_yticklabels()): 
    #ax.get_xticklabels() = all the labels on the x-axis
    #ax.get_yticklabels() = all the labels on the y-axis
    label.set_fontname('Times New Roman')
    
plt.tight_layout()
plt.show()


#-----HEAT MAP-----
sns.heatmap(df_temp, 
            annot=True,         #annotation, number inside the box
            cmap='coolwarm',    #“Cool” colors for low values, “warm” (red-ish) for high values.
            fmt='.1f')          #Format of the annotated numbers. '.1f' = float with 1 decimal place
plt.title("Patient Temperatures Heatmap")
plt.ylabel("Patients")
plt.xlabel("Time of Day")
plt.tight_layout()              #Adjusts spacing between plot elements so they don’t overlap or get cut off
plt.show()
```
